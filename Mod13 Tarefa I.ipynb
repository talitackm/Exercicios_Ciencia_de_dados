### Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.

```python
y, x = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + C(qtd_filhos) + tipo_renda + educacao + estado_civil + tipo_residencia + idade + tempo_emprego + C(qt_pessoas_residencia)', data = df)
x
```

```python
reg = smf.ols('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + C(qtd_filhos) + tipo_renda + educacao + estado_civil + tipo_residencia + idade + tempo_emprego + C(qt_pessoas_residencia)', 
              data = df ).fit()
reg.summary()
```

### Remova a variável menos significante e analise:

```python
reg = smf.ols('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + C(qtd_filhos) + tipo_renda + educacao + estado_civil +idade + tempo_emprego + C(qt_pessoas_residencia)', 
              data = df ).fit()
reg.summary()
```

- ### Com a remoção da variável tipo_residecia o AIC diminuiu, assim houve uma melhora no modelo com a retirada dela

### Siga removendo as variáveis menos significantes, sempre que o p-value for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor.

```python
reg = smf.ols('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + educacao  + tipo_renda + estado_civil +idade + tempo_emprego + C(qt_pessoas_residencia)', 
              data = df ).fit()
reg.summary()
```

- ### Vemos que com a retirada da variável qtd_filhos a AIC continuou a mesma, porém o BIC aumentou e como ele penaliza mais fortemente a complexidade do modelo, com a retirada dessa variável o modelo deve ter ficado mais complexo.

```python
reg = smf.ols('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + estado_civil +idade + tempo_emprego + C(qt_pessoas_residencia)', 
              data = df ).fit()
reg.summary()
```

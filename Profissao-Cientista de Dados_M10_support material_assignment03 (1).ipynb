1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
previsao_renda = pd.read_csv('previsao_de_renda.csv')
previsao_renda.head()
````

````python
ax = sns.countplot(x='data_ref',  hue = 'posse_de_veiculo', data=previsao_renda)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

# tick_labs.tolist()

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'posse_de_imovel', data=previsao_renda)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

# tick_labs.tolist()

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
tab = pd.crosstab(previsao_renda['data_ref'],previsao_renda['sexo'])
tab = tab.div(tab.sum(axis = 1), axis = 0)
tab
````

````python
ax = tab.plot.bar(stacked = True)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'tipo_renda', data=previsao_renda)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90)

# tick_labs.tolist()

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'sexo',
                   data=previsao_renda, dodge=True, errorbar=('ci', 95))


plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90);
````

````python
ax = sns.pointplot(x="data_ref", y="renda", hue = 'posse_de_imovel',
                   data=previsao_renda, dodge=True, errorbar=('ci', 95))


plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)

ax.set_xticklabels(ax.get_xticklabels(),rotation=90);
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
demo_semdup = pd.get_dummies(demo_semdup, columns = ['tipo_renda','estado_civil','tipo_residencia','educacao'], drop_first = True)
````

````python
demo_semdup.head()
````

````python
demo_semdup.posse_de_veiculo = demo_semdup.posse_de_veiculo.map({'Y': 1, 'N':0})
demo_semdup.head()
````

````python
demo_semdup.posse_de_imovel = demo_semdup.posse_de_imovel.map({'Y': 1, 'N':0})
demo_semdup.head()
````

````python
demo_semdup.dtypes
````

````python
demo_semdup.columns
````

````python
demo_semdup.tipo_renda_Pensioner = demo_semdup.tipo_renda_Pensioner.astype(int)
demo_semdup['tipo_renda_State servant'] = demo_semdup['tipo_renda_State servant'].astype(int)                   
demo_semdup['tipo_renda_Student'] = demo_semdup['tipo_renda_Student'].astype(int)                             
demo_semdup.tipo_renda_Working = demo_semdup.tipo_renda_Working.astype(int)                          
demo_semdup.estado_civil_Married = demo_semdup.estado_civil_Married.astype(int)                       
demo_semdup.estado_civil_Separated = demo_semdup.estado_civil_Separated.astype(int)                       
demo_semdup['estado_civil_Single / not married'] = demo_semdup['estado_civil_Single / not married'].astype(int)              
demo_semdup.estado_civil_Widow = demo_semdup.estado_civil_Widow.astype(int)                          
demo_semdup['tipo_residencia_House / apartment'] = demo_semdup['tipo_residencia_House / apartment'].astype(int)             
demo_semdup['tipo_residencia_Municipal apartment'] = demo_semdup['tipo_residencia_Municipal apartment'].astype(int)              
demo_semdup['tipo_residencia_Office apartment'] = demo_semdup['tipo_residencia_Office apartment'].astype(int)        
demo_semdup['tipo_residencia_Rented apartment'] = demo_semdup['tipo_residencia_Rented apartment'].astype(int)              
demo_semdup['tipo_residencia_With parents'] = demo_semdup['tipo_residencia_With parents'].astype(int)           
demo_semdup['educacao_Higher education'] = demo_semdup['educacao_Higher education'].astype(int)                      
demo_semdup['educacao_Incomplete higher'] = demo_semdup['educacao_Incomplete higher'].astype(int)                  
demo_semdup['educacao_Lower secondary'] = demo_semdup['educacao_Lower secondary'].astype(int)                     
demo_semdup['educacao_Secondary / secondary special'] = demo_semdup['educacao_Secondary / secondary special'].astype(int) 
demo_semdup.mau = demo_semdup.mau.astype(int)
````

````python
demo_semdup.sexo = demo_semdup.sexo.map({'M': 1, 'F': 0})
demo_semdup.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
email_mau = pd.crosstab(demo_semdup['possui_email'], demo_semdup['mau'])
email_mau_pct = email_mau.div(email_mau.sum(axis=0), axis = 1) 
email_mau_pct
````

````python
posse_de_veiculo_mau = pd.crosstab(demo_semdup['posse_de_veiculo'], demo_semdup['mau'])
posse_de_veiculo_mau_pct = posse_de_veiculo_mau.div(posse_de_veiculo_mau.sum(axis=0), axis = 1) 
posse_de_veiculo_mau_pct
````

### A variável possui_email é mais podeosa para prever a probabilidade de mau=1, já que quando falamos de mau=1, 93% da base de dados não possuem e-mail.

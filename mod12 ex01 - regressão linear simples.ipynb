### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf
````

````python
tips = sns.load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
reg_tip = smf.ols('tip ~ net_bill', data = tips).fit()
reg_tip.summary()
````

````python
# Gráfico de dispersão
sns.regplot(y = 'tip', x = 'net_bill', data = tips[tips['tip_pct']<.5])

# Gerando dados para a linha dos 10%
x = np.linspace(0,tips['net_bill'].max(),100)
y = .1*x

# Plotando a linha dos 10%
plt.plot(x, y, '-r', label='linha dos 10%')

# Opções estéticas para explicitar o (0,0) no gráfico
plt.xlim(left = 0)
plt.ylim(bottom = 0)
````

````python
pred_tip = reg_tip.fittedvalues
pred_tip
````

### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
reg_tip_pct = smf.ols('tip_pct ~ net_bill', data = tips).fit()
reg_tip_pct.summary()
````

````python
# Gráfico de dispersão
sns.regplot(y = 'tip_pct', x = 'net_bill', data = tips[tips['tip_pct']<.5])
````

Compare os dois gráficos e comente (de forma objetiva e suscinta) a relação entre eles e a diferença do ajuste. Podemos comparar diretamente os $R^2$s?

Apesar do valor da gorjeta aumentar conforme o valor da conta aumenta, a porcentagem dada de gorjeta em relação ao valor da conta diminui. Em relação ao ajuste percebe-se mais variação no segundo gráfico, com pontos mais espaçados da linha traçada pelo modelo.
Não, pois o R-quadrado representa a proporção da variabilidade na variável dependente que pode ser explicada pelo modelo e o mesmo não pode determinar se as estimativas e predições dos coeficientes são tendenciosas, e é por isso que devemos avaliar os gráficos de resíduos, por exemplo.

O R-quadrado não indica se um modelo de regressão é adequado. É possível ter um valor baixo de R-quadrado para um bom modelo ou um valor alto de R-quadrado para um modelo que não se encaixa nos dados!
